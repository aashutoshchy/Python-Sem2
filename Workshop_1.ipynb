Q1. Write a program to prompt the user for hours and rate per hour using input( ) to compute gross pay. Pay the hourly rate for the hours up to 40 and 1.5 times the hourly rate for all hours worked above 40 hours. Use 45 hours and a rate of 10.50 per hour to test the program (the pay should be 498.75).

```python
# Prompting user for hours and rate
hours = float(input("Enter total hours: "))
rate = float(input("Enter the rate per hour (in %): "))

# Conditional statement to check if the hours is more than 40 and calculate gross pay based on that.
if hours <= 40 and hours > 0:
    gross_pay = hours * rate
else:
    over_hours = hours - 40
    gross_pay = (over_hours * rate*1.5) + (40*rate)

# Printing the output
print("Gross Pay is: ", gross_pay)
```

Q2. Write a function that compares two numbers and returns them in increasing order.

```python
# Defining the function with the parameter and returning the value in increasing order

def order_numbers(number1, number2):
    if number2 > number1:
        return number2, number1
    else:
        return number1, number2

# Calling the function with argument and storing as tuple.
smaller, bigger = order_numbers(100, 99)
print(smaller, bigger)
```

1. Write a function fib (n) that prompts the number of terms “n” from the user and displays first “n” terms of Fibonacci sequence iteratively.

```python
# Function Definition

def fib(n):
    a = 0
    b = 1
    print(a)
    print(b)
    for i in range(0, n):
        c = a+b
        print(c)

        a = b
        b = c


# Prompting the user for a number
n = int(input("Enter any number: "))

# Function Call
fib(n)
```

2. Write a function myPow(x, n) that returns 𝑥𝑛, where and n is a non-negative integer. Do not use the ** operator or the math.pow function - use one while loop.

```python
def myPow(x, n):
    idx = 0
    ans = 1;
    while idx < n:
        ans = ans * x
        idx += 1
    return ans


print(myPow(2,3))
```

3. Write a function SumDigits(n) that displays the sum of digits of “n”. Hint: SumDigits(234) will be 2+3+4=9

```python
def sumDigits(n):
    sum = 0
    while n != 0:
        rem = n%10
        sum += rem
        n = n//10
    print(sum)

sumDigits(234)
```

4. Write a function squaredSum that returns the sum of the squares of an array/list of integers such that squaredSum([2,3,4]) gives 29. Write your solution in pure Python to show the logic in your solution, do not use any library functions.

```python
def squaredSum(nums):
    sum = 0
    for num in nums:
        sum += num*num
    return sum

print(squaredSum([2,3,4]))
```

5. Write a function called sumDigits that is given an integer num and returns the sum of the digits of num. Eg. sumDigits(6) gives 21. Do not use any library functions (if there is one).

```python
def sumDigits(n):
    sum = 0
    for i in range(1,n+1):
        sum += i
    print(sum)

sumDigits(6)
```
